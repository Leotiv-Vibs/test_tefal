```python
import random
from collections import Counter

import scipy
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
def generate_data(n_features: int = 10, count_examples: int = 10) -> np.array:
    """
    function for generating data
    :param n_features: count of a columns
    :param count_examples: count of a rows 
    :return: 2d  array with random data
    """
    x = np.array([[i * random.randint(1, 5) for i in range(1, n_features + 1)] for j in range(1, count_examples + 1)])
    return x
```

```python
N_FEATURES = 10
N_SAMPLES = 20

data = generate_data(N_FEATURES,N_SAMPLES) # get the generated data
```

```python
norm_data = MinMaxScaler().fit_transform(data)  # normalize the generated data using the minmaxscaller tool
```

Knowing the inner tools is great
But it's much more interesting to implement it yourself and show your deep understanding

```python
def get_mathematical_expectation(data_col: np.array) -> float:
    mean = 0
    for key, value in dict(Counter(data_col)).items():
        mean += key * (value/data_col.shape[0])
        
    return mean

def get_varience(data_col: np.array) -> float:
    avg = sum(data_col) / len(data_col)
    var = sum((x-avg)**2 for x in data_col) / len(data_col)

    return var

def get_std(data_cal: np.array) -> float:
    
    return get_varience(data_cal) ** (1/2)
```

```python
for i in range(data.shape[1]):
    print(f"Mathematical expectation for {i} features {get_mathematical_expectation(data[:,i])}")
    print(f"Varience for {i} features {get_varience(data[:,i])}")
    print(f"Standard scaller for {i} features {get_std(data[:,i])}")
    print("\n")
```
